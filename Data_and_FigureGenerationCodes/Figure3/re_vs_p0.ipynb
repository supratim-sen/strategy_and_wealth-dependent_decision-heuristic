```python
## Importing data files

import numpy as np
import matplotlib.pyplot as plt

data = np.load('re_vs_p0.npz')

result_array = data['result_array']
result2_array = data['result2_array']
```

```python
## Heatmap for fraction of Cooperators after 20 rounds

re_arr = x=np.arange(0,1.2,0.2)
p0_arr = np.linspace(0.55,1.0,num = 6,endpoint=True)

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
from decimal import Decimal
from matplotlib.pyplot import figure
import numpy as np; np.random.seed(0)
import seaborn as sns

data = (np.mean(result2_array,2))
new_re_arr = np.zeros(len(re_arr))
for i in range(len(re_arr)):
    new_re_arr[i] = ("{:.01e}".format(re_arr[i]))
    
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)
figure(num=None, figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
sns.heatmap(data, cmap="YlGnBu", yticklabels=new_re_arr, xticklabels=[0.55,0.64,0.73,0.82,0.91,1.0], annot=True,  linewidths=.5)
plt.ylabel('re', fontname = 'Arial', fontsize=12)
plt.xlabel('$p_0$', fontname = 'Arial', fontsize=12)
#plt.title(' Cooperator fraction after 20 rnd: N = 50, Init. GINI = 0.2, Runs = 30', fontsize=8, pad='10') 
plt.savefig('re vs p_0_coop.png', bbox_inches='tight', format='png',quality=95,dpi=300)
```
