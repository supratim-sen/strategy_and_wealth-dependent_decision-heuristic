```python
##Importing data files

import numpy as np
import matplotlib.pyplot as plt

data = np.load('extreme_testing_N=50,re=0.0.npz')
data2 = np.load('extreme_testing_N=50,re=0.3.npz')
data3 = np.load('extreme_testing_N=50,re=0.6.npz')
data4 = np.load('extreme_testing_N=50,re=1.0.npz')

deg_array = data['deg_array']
result_array = data['result_array']
avg_wealth_array = data['avg_wealth_array']
co_op_array = data['co_op_array']

deg_array_2 = data2['deg_array']
result_array_2 = data2['result_array']
avg_wealth_array_2 = data2['avg_wealth_array']
co_op_array_2 = data2['co_op_array']

deg_array_3 = data3['deg_array']
result_array_3 = data3['result_array']
avg_wealth_array_3 = data3['avg_wealth_array']
co_op_array_3 = data3['co_op_array']

deg_array_4 = data4['deg_array']
result_array_4 = data4['result_array']
avg_wealth_array_4 = data4['avg_wealth_array']
co_op_array_4 = data4['co_op_array']
```

```python
## Plotting fraction of cooperators vs Rounds

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})

x=np.arange(1,21,1)

y1=((np.mean(co_op_array,1))[1])
z1=((np.std(co_op_array,1))[1])

y2=((np.mean(co_op_array_2,1))[1])
z2=((np.std(co_op_array_2,1))[1])

y3=((np.mean(co_op_array_3,1))[1])
z3=((np.std(co_op_array_3,1))[1])

y4=((np.mean(co_op_array_4,1))[1])
z4=((np.std(co_op_array_4,1))[1])

fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)
plt.ylim(0,1)
plt.xlim(0,20)
plt.yticks(np.arange(0, 1.25, step=0.25), fontname = 'Arial', fontsize=10)
plt.xticks(np.arange(0, 21, step=5), fontname = 'Arial', fontsize=10)
plt.grid(0.5,'major', linestyle='--', linewidth=0.5)

ax.plot(x, y1, '-o',color='lightseagreen', label='re=0.0')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2,color='paleturquoise')

ax.plot(x, y2, '-o',color='slategrey',label='re=0.3')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2,color='darkgrey')

ax.plot(x, y3, '-o',color='darkgoldenrod',label='re=0.6')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2,color='orange')

ax.plot(x, y4, '-o',color='crimson', label='re=1.0')
ax.fill_between(x, y4 - z4, y4 + z4, alpha=0.2,color='hotpink')

#plt.title('N = 50, Runs = 100, Init. GINI = 0.2, $\lambda$ = 0.001, $p_r$ = 1.0, $p_b$ = 1.0, $p_m$ = 1.0, $p_e$ = 0.0, $p_s$ = 0.0', fontsize=7.5)
plt.ylabel('Mean fraction of Cooperators', fontname = 'Arial', fontsize=10)
plt.xlabel('Number of rounds', fontname = 'Arial', fontsize=10)
plt.legend(handlelength=4.0, ncol=2, fontsize=9, loc="upper right")
#plt.savefig('strategy_new,N=50.png',format='png',quality=95,dpi=300) 
```
