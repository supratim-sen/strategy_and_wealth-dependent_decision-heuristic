```python
import numpy as np
import matplotlib.pyplot as plt
data = np.load('Supl-3.npz')
c_ar= data['c_ar']
d_ar = data['d_ar']
eq_ar = data['eq_ar']
```

```python
from matplotlib import rc
from matplotlib.pyplot import figure

rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
params = {'mathtext.default': 'regular' }  

x=np.arange(1,21,1)
y1=((np.mean(c_ar,axis=2))[0][0])
z1=((np.std(c_ar,axis=2))[0][0])
y2=((np.mean(d_ar,axis=2))[0][0])
z2=((np.std(d_ar,axis=2))[0][0])
y3=((np.mean(eq_ar,axis=2))[0][0])
z3=((np.std(eq_ar,axis=2))[0][0])   

fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
plt.ylabel('Nature of local environments', fontname = 'Arial', fontsize=12)
plt.xlabel('Number of rounds', fontname = 'Arial', fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0,60)
plt.xticks(np.arange(0, 21, step=5), fontsize=12)

ax.plot(x, y1, '-o',color='navy', label='Cooperative environment')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2, color='dodgerblue')

ax.plot(x, y2, '--o',color='maroon', linestyle='dashed', label='Selfish environment')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2, color ='peru')

ax.plot(x, y3, '-o',color='darkcyan', label='Neutral environment')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2, color='lightgreen')

plt.grid(1,'major',linestyle='--', linewidth=0.5)
plt.legend(handlelength=3.5,fontsize=9, loc="upper right")
plt.savefig('fig-S3b,re=0.0,p0=0.55.png',format='png',quality=95,dpi=300)
```
