```python
import numpy as np
import matplotlib.pyplot as plt
data = np.load('S2b_strat,wealth_vs_rounds,p0=0.65,M=50.npz')
rc_ar = data['rc_ar']
rd_ar = data['rd_ar']
pc_ar = data['pc_ar']
pd_ar = data['pd_ar']
```

```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
x=np.arange(1,51,1)
y1 = (np.mean(rc_ar,1))[3]
z1 = (np.std(rc_ar,1))[3]

y2 = (np.mean(rd_ar,1))[3]
z2 = (np.std(rd_ar,1))[3]

y3 = (np.mean(pc_ar,1))[3]
z3 = (np.std(pc_ar,1))[3]

y4 = (np.mean(pd_ar,1))[3]
z4 = (np.std(pd_ar,1))[3]

from matplotlib.pyplot import figure
fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)

plt.ylabel('No. of players by wealth and strategy', fontsize=11)
plt.xlabel('Number of rounds', fontsize=11)
plt.yticks(fontsize=11)
plt.ylim(0,30)
plt.xticks(np.arange(0, 51, step=5), fontname = 'Arial', fontsize=11)
plt.grid(0.5,'major',linestyle='--', linewidth=0.5)
ax.plot(x,y1,'-o',markevery=3,label='Rich Cooperator')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2)

ax.plot(x,y2,'--o',color='darkorange',markevery=3,linestyle='dashed',label='Rich Defector')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2)

ax.plot(x,y3,'-o',color='g',markevery=3,label='Poor Cooperator')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2)

ax.plot(x,y4,'--o',color='r',markevery=3,label='Poor Defector')
ax.fill_between(x, y4 - z4, y4 + z4, alpha=0.2)

plt.legend(ncol=2, handlelength=3.5, fontsize=9, loc="upper right")
plt.savefig('new_fig_S2b_re=1.0.png',format='png',quality=95,dpi=300)
```
