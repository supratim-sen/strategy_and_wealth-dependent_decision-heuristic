```python
import numpy as np
import matplotlib.pyplot as plt
data = np.load('S9c_strategy,bonds_vs_rounds,p0=0.65,M=50.npz')

cc_ar= data['cc_ar']
cd_ar = data['cd_ar']
dd_ar = data['dd_ar']
```

```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
x=np.arange(1,51,1)

y1=((np.mean(cc_ar,1))[0])
z1=((np.std(cc_ar,1))[0])

y2=((np.mean(cd_ar,1))[0])
z2=((np.std(cd_ar,1))[0])

y3=((np.mean(dd_ar,1))[0])
z3=((np.std(dd_ar,1))[0])   


from matplotlib.pyplot import figure
fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')

plt.ylabel('Type of bonds, C: Cooperative, D: Defective', fontsize=11)
plt.xlabel('Number of rounds', fontsize=11)
ax.plot(x,y1,'-o',markevery=3,label='C-C')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2)

ax.plot(x,y2,'--o',color='darkorange',markevery=3,linestyle='dashed',label='C-D')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2)

ax.plot(x,y3,'-o',color='r',markevery=3,label='D-D')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2, color='indianred')

plt.ylim(0,500)

plt.yticks(np.arange(0,610,step=100),fontsize=11)
plt.xticks(np.arange(0, 51, step=5), fontname = 'Arial', fontsize=11)
plt.grid(1,'major',linestyle='--', linewidth=0.5)

plt.legend(fontsize=10, loc="upper right")
plt.savefig('new_fig-S9c,re=0.0.png',format='png',quality=95,dpi=300)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      
```
