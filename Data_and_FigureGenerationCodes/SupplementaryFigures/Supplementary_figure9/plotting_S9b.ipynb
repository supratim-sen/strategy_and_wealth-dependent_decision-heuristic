```python
import numpy as np
import matplotlib.pyplot as plt
data = np.load('S9b_environ_vs_rounds,p0=0.65,M=50.npz')
c_ar= data['c_ar']
d_ar = data['d_ar']
eq_ar = data['eq_ar']
```

```python
from matplotlib import rc
from matplotlib.pyplot import figure

rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
params = {'mathtext.default': 'regular' }  

x=np.arange(1,51,1)
y1=((np.mean(c_ar,1))[0])
z1=((np.std(c_ar,1))[0])
y2=((np.mean(d_ar,1))[0])
z2=((np.std(d_ar,1))[0])
y3=((np.mean(eq_ar,1))[0])
z3=((np.std(eq_ar,1))[0])   

fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
#figure(num=None, figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
plt.ylabel('Nature of local environments', fontname = 'Arial', fontsize=11)
plt.xlabel('Number of rounds', fontname = 'Arial', fontsize=11)
plt.yticks(fontsize=11)
plt.ylim(0,60)
plt.xticks(np.arange(0, 51, step=5), fontsize=11)
#plt.title('N = 50, Runs = 30, re = 0.3, $\lambda=0.001$, Initial GINI: 0.2', fontsize=8, pad='10')

ax.plot(x, y1, '-o',color='navy', markevery=3, label='Cooperative environment')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2, color='dodgerblue')

ax.plot(x, y2, '--o',color='maroon', markevery=3, linestyle='dashed', label='Selfish environment')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2, color ='peru')

ax.plot(x, y3, '-o',color='darkcyan', markevery=3, label='Neutral environment')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2, color='lightgreen')

plt.grid(1,'major',linestyle='--', linewidth=0.5)
#plt.legend(handlelength=3.5,fontsize=10, loc='best', bbox_to_anchor=(1.0, 0.65))
plt.legend(handlelength=3.5,fontsize=9, loc="upper right")
plt.savefig('new_fig-S9b,re=0.0.png',format='png',quality=95,dpi=300)
```
