```python
from decimal import Decimal
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np; np.random.seed(0)
import seaborn as sns

data = np.load('Supl-1.npz')
environ_arr = data['environ_arr']
strategy_ar = data['strategy_ar']
wealth_record_arr = data['wealth_record_arr']
```

```python
matrix = environ_arr[0,:,:]
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})

from matplotlib.pyplot import figure
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)

fig, ax = plt.subplots(1,1,figsize=(10, 8),dpi=200,facecolor='w', edgecolor='k')
sns.heatmap(matrix, cmap="Blues", xticklabels=7, yticklabels=7, annot=False,  linewidths=.0, vmin=0, vmax=1, ax=ax)
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=20)
ax.tick_params(labelsize=20)

plt.ylabel('No of rounds', fontsize=20, weight='bold')
plt.xlabel('Nodes', fontsize=20, weight='bold')
plt.savefig('envrionment_p0=0.55.png', bbox_inches='tight', format='png',quality=95,dpi=1000)
```
