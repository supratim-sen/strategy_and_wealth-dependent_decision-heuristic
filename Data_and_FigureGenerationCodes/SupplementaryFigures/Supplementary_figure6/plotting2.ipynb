```python
## Importing data files

import numpy as np
import matplotlib.pyplot as plt

data = np.load('visible_invert_pc.npz')
data2 = np.load('visible_wealth.npz')

deg_array = data['deg_array']
result_array = data['result_array']
avg_wealth_array = data['avg_wealth_array']
co_op_array = data['co_op_array']

deg_array_vis = data2['deg_array']
result_array_vis = data2['result_array']
avg_wealth_array_vis = data2['avg_wealth_array']
co_op_array_vis = data2['co_op_array']
```

```python
## Plotting GINI vs Rounds

x=np.arange(0,21,1)

y1=((np.mean(result_array,1))[0])
z1=((np.std(result_array,1))[0])

y2=((np.mean(result_array,1))[1])
z2=((np.std(result_array,1))[1])

y3=((np.mean(result_array,1))[2])
z3=((np.std(result_array,1))[2])   

y4=((np.mean(result_array_vis,1))[0])
z4=((np.std(result_array_vis,1))[0])

y5=((np.mean(result_array_vis,1))[1])
z5=((np.std(result_array_vis,1))[1])

y6=((np.mean(result_array_vis,1))[2])
z6=((np.std(result_array_vis,1))[2])  

fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
plt.yticks(np.arange(0, 0.6, step=0.1), fontname = 'Arial', fontsize=10)
plt.xticks(np.arange(0, 21, step=5), fontname = 'Arial', fontsize=10)
plt.grid(0.5,'major', linestyle='--', linewidth=0.5)
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)

ax.plot(x, y1, 'd', markersize = 5, color='lightseagreen', linestyle='dashdot', fillstyle='top', label='Invert $p_c$, None')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2,color='paleturquoise')

ax.plot(x, y2, 'd', markersize = 5, color='slategrey', linestyle='dashdot', fillstyle='top', label='Invert $p_c$, Low')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2,color='darkgrey')

ax.plot(x, y3, 'd', markersize = 5, color='darkgoldenrod', linestyle='dashdot', fillstyle='top', label='Invert $p_c$, High')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2,color='orange')

ax.plot(x, y4, '-o', markersize = 5, color='lightseagreen', label='Standard $p_c$, None')
ax.fill_between(x, y4 - z4, y4 + z4, alpha=0.2,color='paleturquoise')

ax.plot(x, y5, '-o', markersize = 5, color='slategrey', label='Standard $p_c$, Low')
ax.fill_between(x, y5 - z5, y5 + z5, alpha=0.2,color='darkgrey')

ax.plot(x, y6, '-o', markersize = 5, color='darkgoldenrod', label='Standard $p_c$, High')
ax.fill_between(x, y6 - z6, y6 + z6, alpha=0.2,color='orange')

plt.ylabel('GINI Coefficient', fontname = 'Arial', fontsize=10)
plt.xlabel('Number of rounds', fontname = 'Arial', fontsize=10)
plt.legend(handlelength=4.0, ncol=2, fontsize=8, loc="upper right")
plt.savefig('comparison_invertpc_GINI.png',format='png',quality=95,dpi=300) 
```

```python
## Plotting fraction of cooperators vs Rounds

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})

x=np.arange(1,21,1)
y1=((np.mean(co_op_array,1))[0])
z1=((np.std(co_op_array,1))[0])
y2=((np.mean(co_op_array,1))[1])
z2=((np.std(co_op_array,1))[1])
y3=((np.mean(co_op_array,1))[2])
z3=((np.std(co_op_array,1))[2])

fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
plt.ylim(0,1)
plt.xlim(0,20)
plt.yticks(np.arange(0, 1.25, step=0.25), fontname = 'Arial', fontsize=10)
plt.xticks(np.arange(0, 21, step=5), fontname = 'Arial', fontsize=10)
plt.grid(0.5,'major', linestyle='--', linewidth=0.5)

ax.plot(x, y1, '-o',color='lightseagreen', label='Initial GINI = 0.0')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2,color='paleturquoise')
ax.plot(x, y2, '-o',color='slategrey',label='Initial GINI = 0.2')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2,color='darkgrey')
ax.plot(x, y3, '-o',color='darkgoldenrod',label='Initial GINI = 0.4')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2,color='orange')

plt.ylabel('Mean fraction of Cooperators', fontname = 'Arial', fontsize=10)
plt.xlabel('Number of rounds', fontname = 'Arial', fontsize=10)
plt.legend(handlelength=4.0, fontsize=8, loc="lower right")
plt.savefig('invertpc_strategy.png',format='png',quality=95,dpi=300) 
```
