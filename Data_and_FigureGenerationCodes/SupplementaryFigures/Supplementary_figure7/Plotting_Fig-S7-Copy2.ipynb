```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load('S7_rewiring_test.npz')
final_res_arr = data['result_array']
final_strat_arr = data['co_op_array']
final_weal_arr = data['avg_wealth_array']
final_deg_arr = data['deg_array']
```

```python
## Plotting GINI vs Rounds

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)

x=np.arange(0,21,1)
y1=((np.mean(final_res_arr,2))[2][0])
z1=((np.std(final_res_arr,2))[2][0])
y2=((np.mean(final_res_arr,2))[2][1])
z2=((np.std(final_res_arr,2))[2][1])
y3=((np.mean(final_res_arr,2))[2][2])
z3=((np.std(final_res_arr,2))[2][2])   
y4=((np.mean(final_res_arr,2))[2][3])
z4=((np.std(final_res_arr,2))[2][3])
y5=((np.mean(final_res_arr,2))[2][4])
z5=((np.std(final_res_arr,2))[2][4])
y6=((np.mean(final_res_arr,2))[2][5])
z6=((np.std(final_res_arr,2))[2][5])
y7=((np.mean(final_res_arr,2))[2][6])
z7=((np.std(final_res_arr,2))[2][6])
y8=((np.mean(final_res_arr,2))[2][7])
z8=((np.std(final_res_arr,2))[2][7])
y9=((np.mean(final_res_arr,2))[2][8])
z9=((np.std(final_res_arr,2))[2][8])


fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
plt.yticks(np.arange(0.13, 0.34, step=0.04), fontname = 'Arial', fontsize=10)
plt.xticks(np.arange(0, 21, step=5), fontname = 'Arial', fontsize=10)
plt.grid(0.5,'major', linestyle='--', linewidth=0.5)
plt.ylim(0.13,0.34)

ax.plot(x, y1, '-o',color='aquamarine', label='$p_e=p_s=0.3,p_r=p_b=p_m=0.6$')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2,color='paleturquoise')
ax.plot(x, y2, '-o',color='slategrey', label='$p_e=p_s=0.3,p_r=p_b=p_m=0.7$')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2,color='darkgrey')
ax.plot(x, y3, '-o',color='darkgoldenrod', label='$p_e=p_s=0.3,p_r=p_b=p_m=0.8$')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2,color='orange')
ax.plot(x, y4, '-o',color='steelblue', label='$p_e=p_s=0.3,p_r=p_b=p_m=0.9$')
ax.fill_between(x, y4 - z4, y4 + z4, alpha=0.2,color='dodgerblue')
ax.plot(x, y5, '-o',color='crimson', label='$p_e=p_s=0.1,p_r=p_b=p_m=0.8$')
ax.fill_between(x, y5 - z5, y5 + z5, alpha=0.2,color='pink')
ax.plot(x, y6, '-o',color='indigo', label='$p_e=p_s=0.2,p_r=p_b=p_m=0.8$')
ax.fill_between(x, y6 - z6, y6 + z6, alpha=0.2,color='mediumpurple')
ax.plot(x, y7, '-o',color='gold', label='$p_e=p_s=0.4,p_r=p_b=p_m=0.8$')
ax.fill_between(x, y7 - z7, y7 + z7, alpha=0.2,color='khaki')
ax.plot(x, y8, '-o',color='darkcyan', label='$p_e=p_s=0.4,p_b=0.6,p_r=p_m=0.7$')
ax.fill_between(x, y8 - z8, y8 + z8, alpha=0.2,color='cadetblue')
#ax.plot(x, y9, '-o',color='black', label='$p_e=0.3,p_s=0.2,p_b=0.7,p_r=0.87,p_m=0.93$')
ax.plot(x, y9, '-o',color='black', label='Standard Case')
ax.fill_between(x, y9 - z9, y9 + z9, alpha=0.2,color='dimgray')

plt.ylabel('GINI Coefficient', fontname = 'Arial', fontsize=10)
plt.xlabel('Number of rounds', fontname = 'Arial', fontsize=10)
plt.legend(handlelength=4.0, ncol=3, fontsize=5, loc="upper right")
plt.savefig('new_S7,re=0.6_GINI.png',format='png',quality=95,dpi=300) 
```
