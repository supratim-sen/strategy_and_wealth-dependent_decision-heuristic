```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load('S7_rewiring_test.npz')
final_res_arr = data['result_array']
final_strat_arr = data['co_op_array']
final_weal_arr = data['avg_wealth_array']
final_deg_arr = data['deg_array']
```

```python
## Plotting GINI vs Rounds

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)

x=np.arange(0,21,1)

# re = 0.0 plots
y1=((np.mean(final_res_arr,2))[0][0])  # expt. case
z1=((np.std(final_res_arr,2))[0][0])
y2=((np.mean(final_res_arr,2))[0][8])  # standard case
z2=((np.std(final_res_arr,2))[0][8])

# re = 0.3 plots
y3=((np.mean(final_res_arr,2))[1][0])  # expt. case
z3=((np.std(final_res_arr,2))[1][0])
y4=((np.mean(final_res_arr,2))[1][8])  # standard case
z4=((np.std(final_res_arr,2))[1][8])

# re = 0.6 plots
y5=((np.mean(final_res_arr,2))[2][0])  # expt. case
z5=((np.std(final_res_arr,2))[2][0])
y6=((np.mean(final_res_arr,2))[2][8])  # standard case
z6=((np.std(final_res_arr,2))[2][8])


fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
plt.yticks(np.arange(0.1, 0.4, step=0.05), fontname = 'Arial', fontsize=11)
plt.xticks(np.arange(0, 21, step=5), fontname = 'Arial', fontsize=11)
plt.grid(0.5,'major', linestyle='--', linewidth=0.5)
plt.ylim(0.12,0.34)

# Plotting the experimental curves
ax.plot(x, y1, '-o',color='aquamarine', linestyle='dashed', label='variation, re = 0.0')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2,color='paleturquoise')                              
ax.plot(x, y3, '-p',color='aquamarine', linestyle='dashed', label='variation, re = 0.3')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2,color='paleturquoise')
ax.plot(x, y5, '-*',color='aquamarine', linestyle='dashed', label='variation, re = 0.6')
ax.fill_between(x, y5 - z5, y5 + z5, alpha=0.2,color='paleturquoise')

# Plotting the standard curves
ax.plot(x, y2, '-o',color='black', label='Standard, re = 0.0')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2,color='dimgray')
ax.plot(x, y4, '-p',color='black', label='Standard, re = 0.3')
ax.fill_between(x, y4 - z4, y4 + z4, alpha=0.2,color='dimgray')                           
ax.plot(x, y6, '-*',color='black', label='Standard, re = 0.6')
ax.fill_between(x, y6 - z6, y6 + z6, alpha=0.2,color='dimgray')

plt.ylabel('GINI Coefficient', fontname = 'Arial', fontsize=11)
plt.xlabel('Number of rounds', fontname = 'Arial', fontsize=11)
plt.legend(handlelength=4.0, ncol=2, fontsize=8, loc="upper right")
plt.savefig('S7d_GINI.png',format='png',quality=95,dpi=300) 
```
