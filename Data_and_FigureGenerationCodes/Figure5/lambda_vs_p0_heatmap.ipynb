```python
## Importing data files

import numpy as np
import matplotlib.pyplot as plt

data = np.load('lam_vs_p0.npz')
result_array = data['result_array']
```

```python
## Heatmap for GINI Coefficient after 20 rounds

lam_arr = np.logspace(-6,0,num=8,endpoint=True)
new_lam_arr = np.zeros(len(lam_arr))
for i in range(len(lam_arr)):
    new_lam_arr[i] = ("{:.01e}".format(lam_arr[i]))
p0_arr = np.linspace(0.55,1.0,num = 6,endpoint=True)

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
from decimal import Decimal
from matplotlib.pyplot import figure
import numpy as np; np.random.seed(0)
import seaborn as sns

data = (np.mean(result_array,2))
figure(num=None, figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
ax = sns.heatmap(data, cmap=sns.cm.rocket_r, yticklabels=new_lam_arr, xticklabels=[0.55,0.64,0.73,0.82,0.91,1.0], annot=True,  linewidths=.5)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)

plt.setp(ax.get_xticklabels(), fontsize=24)
plt.setp(ax.get_yticklabels(), fontsize=24)
plt.ylabel('$\lambda$', fontname = 'Arial', fontsize=20)
plt.xlabel('$p_0$', fontname = 'Arial', fontsize=20)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
plt.tight_layout()
plt.savefig('lam_vs_p_0_GINI.png', bbox_inches='tight', format='png',quality=95,dpi=300)
```
