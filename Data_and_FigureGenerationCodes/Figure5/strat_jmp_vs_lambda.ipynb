```python
## Importing data file

import numpy as np
import matplotlib.pyplot as plt

data = np.load('strat_jmp_vs_lambda.npz')

cd_ar = data['cd_ar']
dc_ar = data['dc_ar']
```

```python
## Plotting Strategy-jumps vs lambda

lam_arr = np.logspace(-6,0,num=15,endpoint=True)
x=np.log10(lam_arr)
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})

y2=np.mean(cd_ar,1)
z2=np.std(cd_ar,1)

y3=np.mean(dc_ar,1)
z3=np.std(dc_ar,1)  

from matplotlib.pyplot import figure
fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)

plt.ylabel('Average # strategy shifts', fontsize=12)
plt.xlabel('log$_{10}\lambda$', fontname = 'Arial', fontsize=12)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)

ax.plot(x,y2,'--o',linestyle='dashed',label='C-D')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2)

ax.plot(x,y3,'-o',label='D-C')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2)

plt.grid(1,'major',linestyle='--', linewidth=0.5)
plt.legend(handlelength=3.5, fontsize=10, loc="upper right")
plt.savefig('strat_jmps_vs_lambda.png',format='png',quality=95,dpi=300)
```
