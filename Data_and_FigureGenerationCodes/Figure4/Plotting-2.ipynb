```python
## Importing data file

import numpy as np
import matplotlib.pyplot as plt

data = np.load('degree_vs_re,p0=0.7_new.npz')

cc_deg_ar = data['cc_deg_ar']
cd_deg_ar = data['cd_deg_ar']
dc_deg_ar = data['dc_deg_ar']
dd_deg_ar = data['dd_deg_ar']
ce_deg_ar = data['ce_deg_ar']
de_deg_ar = data['de_deg_ar']
```

```python
## Plotting Mean Degree vs re

re_ar = np.arange(0,1.1,0.1)
x=re_ar
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})

y1=((np.mean(cc_deg_ar,1)))
z1=((np.std(cc_deg_ar,1)))

y2=((np.mean(cd_deg_ar,1)))
z2=((np.std(cd_deg_ar,1)))

y3=((np.mean(dc_deg_ar,1)))
z3=((np.std(dc_deg_ar,1)))

y4=((np.mean(dd_deg_ar,1)))
z4=((np.std(dd_deg_ar,1)))

y5=((np.mean(ce_deg_ar,1)))
z5=((np.std(ce_deg_ar,1)))

y6=((np.mean(de_deg_ar,1)))
z6=((np.std(de_deg_ar,1)))

from matplotlib.pyplot import figure
fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')

plt.ylabel('Average degree in \n different strategy environments', fontsize=16)
plt.xlabel('re', fontsize=20)
plt.yticks(fontsize=16)
plt.xticks(np.arange(0.0, 1.1, step=.1), fontsize=15)
plt.grid(1,'major',linestyle='--', linewidth=0.5)

ax.plot(x,y1,'-o',label='C in Cooperative')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2)

ax.plot(x,y2,'-o',label='C in Selfish')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2)

ax.plot(x,y5,'-o',label='C in Neutral')
ax.fill_between(x, y5 - z5, y5 + z5, alpha=0.2)

ax.plot(x,y3,'--o',color='mediumpurple',linestyle='dashed',label='D in Cooperative')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2)

ax.plot(x,y4,'--o',color='r',linestyle='dashed',label='D in Selfish')
ax.fill_between(x, y4 - z4, y4 + z4, alpha=0.2)

ax.plot(x,y6,'--o',color='brown',linestyle='dashed',label='D in Neutral')
ax.fill_between(x, y6 - z6, y6 + z6, alpha=0.2)

plt.legend(ncol=2, handlelength=3.5, fontsize=11, loc="best",framealpha=0)
plt.tight_layout()
plt.savefig('Degree_vs_re.png',format='png',quality=95,dpi=300)
```
