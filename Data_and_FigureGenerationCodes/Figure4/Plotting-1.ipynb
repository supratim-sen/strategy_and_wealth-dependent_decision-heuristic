```python
## Importing data file

import numpy as np
import matplotlib.pyplot as plt

data = np.load('strat_wealth_vs_re.npz')

rc_ar = data['rc_ar']
pd_ar = data['pd_ar']
rd_ar = data['rd_ar']
pc_ar = data['pc_ar']

c_env_ar = data['c_env_ar']
d_env_ar = data['d_env_ar']
eq_env_ar = data['eq_env_ar']
coop_ar = data['coop_ar']
def_ar = data['def_ar']
```

```python
## Plotting Wealth+Strategy vs re

re_ar = np.arange(0,1.1,0.1)
x=re_ar
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})

y1=np.mean(rc_ar,1)
z1=np.std(rc_ar,1)

y2=np.mean(pd_ar,1)
z2=np.std(pd_ar,1)

y3=np.mean(rd_ar,1)
z3=np.std(rd_ar,1)

y4=np.mean(pc_ar,1)
z4=np.std(pc_ar,1)

from matplotlib.pyplot import figure

fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')

plt.ylabel('Cumulative # strategies \n over last 19 rounds', fontsize=16)
plt.xlabel('re', fontsize=20)
plt.yticks(fontsize=16)
plt.xticks(np.arange(0.0, 1.1, step=.1), fontsize=16)
plt.xticks(fontsize=16)

#plt.title('N = 50, Runs = 50, $\lambda=0.001$, Initial GINI: 0.2', fontsize=8, pad='10')
ax.plot(x,y1,'-o',label='Rich Cooperator')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2)

ax.plot(x,y3,'--o',color='darkorange',linestyle='dashed',label='Rich Defector')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2)

ax.plot(x,y4,'-o',color='g',label='Poor Cooperator')
ax.fill_between(x, y4 - z4, y4 + z4, alpha=0.2)

ax.plot(x,y2,'--o',color='r',linestyle='dashed',label='Poor Defector')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2)

plt.grid(1,'major',linestyle='--', linewidth=0.5)
plt.legend(ncol=2,handlelength=3.5, fontsize=10.5, loc="upper left",bbox_to_anchor=(0.17,0.66),framealpha=0)
plt.tight_layout()
plt.savefig('strategy_vs_re1.png',format='png',quality=100,dpi=300)
```

```python
## Plotting Environment+Strategy vs re

re_ar = np.arange(0,1.1,0.1)
x=re_ar
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})

y1=np.mean(c_env_ar,1)
z1=np.std(c_env_ar,1)

y2=np.mean(d_env_ar,1)
z2=np.std(d_env_ar,1)

y3=np.mean(eq_env_ar,1)
z3=np.std(eq_env_ar,1)

y4=np.mean(coop_ar,1)
z4=np.std(coop_ar,1)

y5=np.mean(def_ar,1)
z5=np.std(def_ar,1)

from matplotlib.pyplot import figure

fig,ax = plt.subplots(figsize=(6, 4), dpi=600, facecolor='w', edgecolor='k')

plt.ylim(0,1000)
plt.ylabel('Cumulative # strategies & \n # strategy environments in 20 rounds', fontsize=14)
plt.xlabel('re', fontsize=20)
plt.yticks(fontsize=16)
plt.xticks(np.arange(0.0, 1.1, step=.1), fontsize=15)

#plt.title('N = 50, Runs = 50, $\lambda=0.001$, Initial GINI: 0.2', fontsize=8, pad='10')
ax.plot(x, y1, '-o',color='navy', label='Cooperative environment')
ax.fill_between(x, y1 - z1, y1 + z1, alpha=0.2, color='dodgerblue')

ax.plot(x, y2, '--o',color='maroon', linestyle='dashed', label='Selfish environment')
ax.fill_between(x, y2 - z2, y2 + z2, alpha=0.2, color ='peru')

ax.plot(x, y3, '-o',color='darkcyan', label='Neutral environment')
ax.fill_between(x, y3 - z3, y3 + z3, alpha=0.2, color='lightgreen')

ax.plot(x, y4, '-o',color='blue', label='Cooperator')
ax.fill_between(x, y4 - z4, y4 + z4, alpha=0.2, color='darkturquoise')

ax.plot(x, y5, '--o',color='r', linestyle='dashed', label='Defector')
ax.fill_between(x, y5 - z5, y5 + z5, alpha=0.2, color='crimson')

plt.grid(1,'major',linestyle='--', linewidth=0.5)
plt.legend(ncol=2, handlelength=0.5, fontsize=10.5,loc="upper left",bbox_to_anchor=(0.3,0.81),framealpha=0)
plt.tight_layout()
plt.savefig('environment_vs_re1.png',format='png',quality=100,dpi=300)
```
